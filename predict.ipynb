# Inbound Prediction

data process

```python
# 總資料
data = pd.read_excel('data.xlsx')
```

```python
# 預測資料
Year= int(input("預測年份："))
Month = int(input("預測月份："))

data_predict = data[(data['Year'] == Year) & (data['Month'] == Month)]
data_predict = data_predict.drop(['Inbound'], axis=1).reset_index(drop = True)
```

```python
# 訓練資料
data_train = data[data['Inbound'].notnull()]
```

```python
# https://stackoverflow.com/questions/50733014/linear-regression-with-dummy-categorical-variables
# 很好用
```

# Model

```python
from statsmodels.formula.api import ols
```

```python
model = ols('Inbound ~ C(Year) + C(Month) + C(Weekday) + C(Bank) + C(Credit) + C(Vacation) + C(Period) + C(ChineseNY)', data = data_train).fit() 
```

```python
model.summary()
```

```python
# 預測
```

```python
predictions = model.predict(data_predict).round()
```

# Output

```python
data_predict['Prediction'] = predictions
data_predict
```

```python
# 抓取Simulation資料進來
simulation = pd.read_csv('simulation.csv', encoding = 'utf-8')
simulation
```

```python
time = simulation['Time'].tolist()
```

```python
# Count Mon - Sun
day = ['Mon-Inbound', 'Tue-Inbound', 'Wed-Inbound', 'Thur-Inbound', 'Fri-Inbound', 'Fri-Inbound', 'Fri-Inbound']
col = simulation.columns.tolist()
count = 0
for i in range(0, len(col)):
    for j in range(0, len(day)):
        
        if col[i] == day[j]:
            day[j] = simulation[col[i]].sum()
day
```

```python
# 找乘數
a = data_predict['Year'].tolist()
mode = []
for i in range(0, len(a)):
    for j in range(0, len(day)):
        if data_predict['Weekday'][i] - 1 == j:
            result = data_predict['Prediction'][i] / day[j]
            mode.append(result)
data_predict['Mode'] = mode
data_predict
```

```python
# Need's columns
a = data_predict['Day'].tolist()

b = data_predict['Day'].astype(str).tolist()
```

```python
col = data_predict['Day'].tolist()
col.insert(0, '')

coln = data_predict['Day'].tolist()
coln.insert(0, '')
col
```

```python
for i in range(0, len(a)):
    weekday = (data_predict['Weekday'][data_predict['Day'] == a[i]]).tolist()[0]
    if weekday == 1:
        b[i] = simulation['Mon'].tolist()
        
    if weekday == 2:
        b[i] = simulation['Tue'].tolist()
    if weekday == 3:
        b[i] = simulation['Wed'].tolist()
    if weekday == 4:
        b[i] = simulation['Thur'].tolist()
    if weekday == 5:
        b[i] = simulation['Fri'].tolist()
    if weekday == 6:
        b[i] = simulation['Fri'].tolist()
    if weekday == 7:
        b[i] = simulation['Fri'].tolist()
    for j in range(0, len(b[i])):
        b[i][j] = round(int(b[i][j]) * data_predict['Mode'][data_predict['Day'] == a[i]].tolist()[0] / 0.9)

```

```python
for i in range(0, len(col)):
    if i == 0:
        col[i] = simulation['Time'].tolist()
    else:
        col[i] = b[i - 1]
col
```

```python
# Output dataframe

Need = pd.DataFrame(data = col)

Ne = Need.T
Ne.columns = coln

```

```python
Ne.to_csv('Need.csv', index = 0, encoding = 'utf-8_sig')
```
